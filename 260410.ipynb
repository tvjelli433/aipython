```python
fruits = ["apple", "banana", "strawberry", "mango", "melon"]
```

```python
for f in fruits:
    print(f)
```

```python
upperf = []
for f in fruits:
  upperf.append(f.upper())
print(upperf)
```

```python
for i, fruit in enumerate(fruits, start=1): #start=1은 1. 부터 나오게 하는 역할. start 없이도 i+1 하면 똑같이 나옴.
    print(f"{i}. {fruit}")
```

```python
joseon_kings = [
    "태조", "정종", "태종", "세종", "문종", "단종", "세조", "예종", "성종",
    "연산군", "중종", "인종", "명종", "선조", "광해군", "인조", "효종", "현종",
    "숙종", "경종", "영조", "정조", "순조", "헌종", "철종", "고종", "순종"
]
```

```python
# (1) 숫자(몇 대) + 왕 이름 목록 출력
for i, king in enumerate(joseon_kings, start=1):
    print(f"{i}대: {king}")
```

```python
# (2) n대를 물어보면 답해주는 기능
n = int(input("몇 대 왕을 찾을까요? : "))

if 1 <= n <= len(joseon_kings): #n이 1~27 사이의 수인지
    print(f"조선 {n}대 왕은 {joseon_kings[n-1]}입니다.") #리스트는 0부터 시작하기 때문에 -1 해줌
else:
    print(f"1부터 {len(joseon_kings)} 사이 숫자를 입력해 주세요.")
```

```python
name = input("왕 이름을 입력하세요 (예: 세종) : ").strip()

if name in joseon_kings: #name이 리스트에 있으면
    n = joseon_kings.index(name) + 1  # 왕 리스트에서 해당 name이 몇 번째인지 => index는 0부터 시작이므로 +1
    print(f"{name}은(는) 조선 {n}대 왕입니다.")
else:
    print(f"'{name}'은(는) 목록에 없습니다.")
```

```python
for i, king in enumerate(reversed(joseon_kings)):
    n = len(joseon_kings) - i #27에서 i를 빼면 몇 대인지 나옴 (ex.1번째 i에서는 27번째 왕이 나옴)
    print(f"{n}대: {king}")
```
