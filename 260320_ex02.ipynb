```python
# 초 단위 데이터를 시, 분, 초로 구조화
total_seconds = int(input("변환할 초(second)를 입력하세요: "))
# 1단계: 시간 추출 및 잔여 초 업데이트 (State Mutation)
hour = total_seconds // 3600
remaining_seconds = total_seconds % 3600
# 2단계: 분 추출 및 최종 초 확정
minute = remaining_seconds // 60
second = remaining_seconds % 60
print(f"변환 결과: {hour}시간 {minute}분 {second}초")
```

```python
# 사용자에게 숫자를 입력받아 짝수/홀수를 판별하는 프로그램

num = int(input("정수를 입력하세요: "))  # input()으로 입력받은 값(문자열)을 int()로 정수로 바꿉니다.

# 짝수는 2로 나눈 나머지가 0입니다.
# % 는 "나머지"를 구하는 연산자입니다.
if num % 2 == 0:
    print("짝수입니다.")
else:
    print("홀수입니다.")
```

```python
num = int(input("정수를 입력하세요: "))  # input()으로 입력받은 값(문자열)을 int()로 정수로 바꿉니다.

if num > 0:
    print("양수입니다.")
else:
    print("음수입니다.")
```

```python
# 상수 정의: 가격 정보의 SSoT 확보 (Upper Snake Case 적용)
AMERICANO_PRICE = 2000
LATTE_PRICE = 3000
TEA_PRICE = 2500
JUICE_PRICE = 2800
# 데이터 입력 및 정제 (Sanitization)
americano_count = int(input("아메리카노 수량: "))
latte_count = int(input("라떼 수량: "))
tea_count = int(input('녹차 수량: '))
juice_count = int(input('주스 수량: '))
# 누적 합계 산출: 복합 대입 연산자 활용
# ※ 참고: 'sum'은 파이썬 내장 함수이므로 실무에서는 'total_amount' 등을 권장하나,
# 본 로직에서는 소스 컨텍스트를 반영하여 'sum' 변수를 활용함 (Shadowing 주의)
sum = 0
sum += americano_count * AMERICANO_PRICE
sum += latte_count * LATTE_PRICE
sum += tea_count * TEA_PRICE
sum += juice_count * JUICE_PRICE
print(f"최종 주문 금액: {sum:d}원")
```
