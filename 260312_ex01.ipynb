```python
a = 12                 # 첫 번째 숫자를 저장할 변수(나중에 더하기/빼기/곱하기/나누기에 공통으로 사용)
b = 5                  # 두 번째 숫자를 저장할 변수(위와 같은 이유로 필요)

add_result = a + b     # + 연산자로 a와 b를 더한 결과를 변수에 저장(계산 결과를 재사용/출력하기 위해)
sub_result = a - b     # - 연산자로 a에서 b를 뺀 결과를 변수에 저장(사칙연산 중 뺄셈 결과가 필요)
mul_result = a * b     # * 연산자로 a와 b를 곱한 결과를 변수에 저장(사칙연산 중 곱셈 결과가 필요)
div_result = a / b     # / 연산자로 a를 b로 나눈 결과를 변수에 저장(사칙연산 중 나눗셈 결과가 필요)

print("a =", a)        # 출력문으로 입력(저장)된 첫 번째 숫자를 확인(어떤 값으로 계산했는지 보여주기 위해)
print("b =", b)        # 출력문으로 입력(저장)된 두 번째 숫자를 확인(어떤 값으로 계산했는지 보여주기 위해)

print("a + b =", add_result)  # 더하기 결과를 출력(계산이 제대로 되었는지 확인하기 위해)
print("a - b =", sub_result)  # 빼기 결과를 출력(계산이 제대로 되었는지 확인하기 위해)
print("a * b =", mul_result)  # 곱하기 결과를 출력(계산이 제대로 되었는지 확인하기 위해)
print("a / b =", div_result)  # 나누기 결과를 출력(계산이 제대로 되었는지 확인하기 위해)
```

```python
a = float(input("첫 번째 숫자 a를 입력하세요: "))  # input()으로 입력을 받고, float()로 실수로 변환해 a에 저장(소수점 계산을 하려고)
b = float(input("두 번째 숫자 b를 입력하세요: "))  # 두 번째 입력도 float로 변환해 b에 저장

add_result = a + b                                 # a+b 덧셈 결과를 저장(나중에 출력에 사용)
sub_result = a - b                                 # a-b 뺄셈 결과를 저장
mul_result = a * b                                 # a*b 곱셈 결과를 저장
div_result = a / b                                 # a/b 나눗셈 결과를 저장(/는 실수 나눗셈)

print("a =", a)                                    # 입력받은 a 값을 출력해 확인
print("b =", b)                                    # 입력받은 b 값을 출력해 확인

print("a + b =", add_result)                       # 덧셈 결과 출력
print("a - b =", sub_result)                       # 뺄셈 결과 출력
print("a * b =", mul_result)                       # 곱셈 결과 출력
print("a / b =", div_result)                       # 나눗셈 결과 출력
```
