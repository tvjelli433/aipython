```python
# pop 함수 활용 예시
prices = {'아메리카노': 2000, '라떼': 3000, '녹차': 2500}
a = prices.pop('라떼') # '라떼' 항목을 삭제하고, 가격인 3000을 a에 저장
print(a) # 출력: 3000
print(prices) # 출력: {'아메리카노': 2000, '녹차': 2500}
```

```python
prices = {'아메리카노': 2000, '라떼': 3000, '녹차': 2500}

# 1. keys() 활용: 이름표만 확인
for key in prices.keys():
 print(key, end=' ') # 아메리카노 라떼 녹차
print('')

# 2. values() 활용: 가격들만 확인
for val in prices.values():
 print(val) # 2000 \n 3000 \n 2500

# 3. items() 활용: 이름과 가격을 동시에!
for key, value in prices.items(): #키와 값 두개 다 써주기
 print(f"{key}의 가격은 {value}원입니다.")
```

1. "키(Key)는 지문과 같다. 중복되지 않도록 유일하게 관리하라."
2. "데이터를 찾을 때는 번호가 아닌 이름(Key)으로 당당하게 요구하라."
3. "키와 값을 함께 다룰 때는 items() 함수가 가장 빠른 지름길이다.

```python
cafe_menu = {
    "아메리카노": 3000,
    "카페라떼": 3500,
    "카푸치노": 4000,
    "바닐라라떼": 4500,
    "카라멜마끼아또": 4800,
    "카페모카": 4500,
    "에스프레소": 2500,
    "녹차라떼": 4300,
    "초코라떼": 4200,
    "아이스티": 3500
}

#각 메뉴 가격 출력하기
sum = 0

for key, value in cafe_menu.items():
 print(f"{key}의 가격은 {value}원입니다.")

#모든 메뉴의 가격 합 출력하기
for value in cafe_menu.values():
  sum = sum + value
print((f"총 가격은 {sum}원입니다."))

#주문 받고 총 주문금액 출력하기
total = 0

while True: #무한반복문. 주문 횟수를 정하지 않았기 때문에 While을 사용함.
    order = input("주문할 메뉴를 입력하세요 (주문 종료 시 \'종료\' 입력): ").strip() #str(input())이나 input()이나 같음. 원래 input은 문자열로 받기 때문임. .strip은 앞뒤 공백 없애는 기능

    if order == "종료":
        break #'종료' 입력시 반복 종료

    if order not in cafe_menu: #order이 메뉴에 없으면
        print("해당 메뉴는 없습니다. 오타를 확인해주세요.")

    else:
        n = int(input(f"{order}의 수량을 입력하세요: ")) #정수값을 받기
        price = cafe_menu[order] * n #cafe_menu[order]는 입력한 메뉴(키)의 가격(값)임.
        total += price
        print(f"{order} {n}잔이 추가되었습니다.")

print(f"최종 결제 금액은 {total}원입니다.")
```
