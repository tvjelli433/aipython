```python
h1=float(input("학생1의 키를 적어주세요:"))
h2=float(input("학생2의 키를 적어주세요:"))

if h1 > h2:
  print('키 차이:', h1 - h2)
else:
  print('키 차이:', h2 - h1)
```

```python
s=float(input('점수를 입력하세요:'))

if s >= 95:
  print("A+")
elif s >= 90:
  print("A0")
elif s >= 85:
  print("B+")
elif s >= 80:
  print("B0")
elif s >= 75:
  print("C+")
elif s >= 70:
  print("C0")
else:
  print("D")
print(s)
```

```python
import random

r1=random.randint(1,3)
r2=random.randint(1,3)
#1가위 2바위 3보

if r1==1:
  print("사람1: 가위")
elif r1==2:
  print("사람1: 바위")
else:
  print("사람1: 보")

if r2==1:
  print("사람2: 가위")
elif r2==2:
  print("사람2: 바위")
else:
  print("사람2: 보")
#if, elif만 조건식임. else는 조건문이 없음.
if (r1==1 and r2==2) or (r1==2 and r2==3) or (r1==3 and r2==1):
  print("사람 2가 이겼습니다")
elif r1==r2:
  print("비겼습니다")
else:
  print("사람 1이 이겼습니다")
```
