```python
s=int(input('점수를 입력하세요:'))

if s > 100 or s < 0:
  print("점수는 0~100 사이여야 합니다")
elif s >= 90:
  print("A")
elif s >= 80:
  print("B")
elif s >= 70:
  print("C")
elif s >= 60:
  print("D")
else:
  print("F")
```

```python
count=0
import random
answer=random.randint(1,100)

user=int(input("추측값을 입력하세요:"))

while user != answer:
  if user > answer:
      print("더 작은 수를 입력하세요.")
      count+=1
      user=int(input("추측값을 입력하세요:"))
  else:
      print("더 큰 수를 입력하세요.")
      count+=1
      user=int(input("추측값을 입력하세요:"))

print("정답입니다.")
count+=1
print("시도횟수:", count)
```

```python
num=int(input("약수를 알고 싶은 숫자를 입력하세요: "))
total=0

for i in range(1, num+1):
  if num%i==0:
    print(i)
    i+=1
    total+=1
  else:
    i+=1

print("약수의 총 개수:", total)
```
